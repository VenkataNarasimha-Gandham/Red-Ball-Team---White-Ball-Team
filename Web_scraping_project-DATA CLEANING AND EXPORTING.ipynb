# Data Cleaning(ODI Matches)

```python
import re
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ODI_batting=pd.read_csv(r"C:\Users\hp\Desktop\Redball-Whiteball\ODI_Batting.csv")
ODI_bowling=pd.read_csv(r"C:\Users\hp\Desktop\Redball-Whiteball\ODI_Bowling.csv")
```

```python
ODI_batting.Out_Not=ODI_batting.Out_Not.apply(lambda x: "Out" if x!="not out " else x)
print(ODI_batting.Out_Not.value_counts())
```

```python
I_bat=list(ODI_batting.Player_name)
Best_10_bat=[]
print(len(I_bat))
for i in range (len(I_bat)):
    if I_bat.count(I_bat[i])>15:
        Best_10_bat.append(I_bat[i])
print(len(Best_10_bat))
ODI_batting["Player_name"]=ODI_batting.Player_name.apply(lambda x: np.nan if x not in Best_10_bat else x)
ODI_batting.dropna(how="any",inplace=True,axis=0)
```

```python
ODI_batting.info()
```

```python
ODI_batting["Time"]=ODI_batting.Time.apply(lambda x: np.nan if x=="-" else x).astype(float)
ODI_batting["Strike_rate"]=ODI_batting.Strike_rate.apply(lambda x: np.nan if x=="-" else x).astype(float)
ODI_batting.info()
```

```python
ODI_batting.Player_name.value_counts()
```

```python
ODI_batting.Player_name=ODI_batting.Player_name.apply(lambda x : x.replace("\xa0",""))
ODI_batting.Player_name=ODI_batting.Player_name.apply(lambda x : x.replace("(c)",""))
ODI_batting.Player_name=ODI_batting.Player_name.apply(lambda x : x.replace("†",""))
```

```python
ODI_batting.describe()
```

```python
ODI_batting.to_csv(r"C:\Users\hp\Desktop\Redball-Whiteball\ODI_Batting_cleaned.csv", index=False)
```

## ** Preparing DataFrames for Exploratory Data Analysis (EDA)**

```python
ODI_batting=pd.read_csv(r"C:\Users\hp\Desktop\Redball-Whiteball\ODI_Batting_cleaned.csv")
ODI_bowling=pd.read_csv(r"C:\Users\hp\Desktop\Redball-Whiteball\ODI_Bowling_cleaned.csv")
```

```python
ODI_batting.head()
```

```python
ODI_bowling.tail()
```

```python
ODI_batting_sorted=ODI_batting.sort_values(by= "Runs",ascending=False)
name=ODI_batting_sorted.Player_name.unique()
group_bat=ODI_batting_sorted.groupby(ODI_batting_sorted.Player_name)
vk=group_bat.get_group("Virat Kohli")
EDA_ODI_bat=pd.concat([(vk.iloc[[1,2,3]]),(vk.iloc[[-3,-2,-1]]),(vk.iloc[[(len(vk)//2)-1,(len(vk)//2),(len(vk)//2)+1]])])
EDA_ODI_bat["total_runs"]=EDA_ODI_bat.Runs.sum()
for i in name:
    ply=group_bat.get_group(i)
    df=pd.concat([(ply.iloc[[1,2,3]]),(ply.iloc[[-3,-2,-1]]),(ply.iloc[[(len(ply)//2)-1,(len(ply)//2),(len(ply)//2)+1]])])
    df["total_runs"]=df.Runs.sum()
    EDA_ODI_bat=pd.concat([EDA_ODI_bat,df])
EDA_ODI_bat.drop_duplicates(inplace=True)
EDA_ODI_bat.duplicated().sum()
EDA_ODI_bat
```

```python
ODI_bowling_sorted=ODI_bowling.sort_values(by= "ECON",ascending=False)
name=ODI_bowling_sorted.Bowler.unique()
group_bowl=ODI_bowling_sorted.groupby(ODI_bowling_sorted.Bowler)
hp=group_bowl.get_group("Hardik Pandya")
EDA_ODI_bowl=pd.concat([(hp.iloc[[1,2,3]]),(hp.iloc[[-3,-2,-1]]),(hp.iloc[[(len(hp)//2)-1,(len(hp)//2),(len(hp)//2)+1]])])
EDA_ODI_bowl["total_wickets"]=EDA_ODI_bowl.Wickets.sum()
for i in name:
    ply=group_bowl.get_group(i)
    df=pd.concat([(ply.iloc[[1,2,3]]),(ply.iloc[[-3,-2,-1]]),(ply.iloc[[(len(ply)//2)-1,(len(ply)//2),(len(ply)//2)+1]])])
    df["total_wickets"]=df.Wickets.sum()
    EDA_ODI_bowl=pd.concat([EDA_ODI_bowl,df])
EDA_ODI_bowl.drop_duplicates(inplace=True)
EDA_ODI_bowl.duplicated().sum()
EDA_ODI_bowl
```
